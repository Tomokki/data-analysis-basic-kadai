```python
# matplotlib.pyplotメソッドとnumpyをインポートするコードを記述
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 折れ線グラフ
y0 = [1, 2, -5, 2]
# nrows=2, ncols=3で、6領域のサブプロットを作成
fig, axes = plt.subplots(2, 3, tight_layout=True)

axes[0, 0].plot(y0)

# sin関数
x = np.linspace(0, 10, 100)
y = 2 + 2 * np.sin(2 * x)

axes[0, 1].plot(x, y, linewidth=2.0)
axes[0, 1].set(xlim=(0, 10), xticks=np.arange(0, 11, 2),
               ylim=(-1, 5), yticks=np.arange(-1, 6, 1))

# ヒストグラム
y_hist = np.random .normal(15, 5, 2000)

axes[0, 2].hist(y_hist, bins=15)

# 散布図
np.random.seed(3)
x_scatter = 4 + np.random.normal(0, 2, 60)
y_scatter = 4 + np.random.normal(0, 2, len(x_scatter))

axes[1, 0].scatter(x_scatter, y_scatter)

plt.tight_layout()
plt.show()
```
